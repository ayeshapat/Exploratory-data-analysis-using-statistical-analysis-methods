```python
#defining a function to perform bootstrapping on a data and return a list of sample means and bootstrapped mean

# n = sample size
# number_iter = number of iterations (1000 in our case)
# data = dataset for resampling
def bootstrap(n,number_iter,data):
  samples=[] #empty list to store random samples from dataset
  sample_means=[] #empty list to store mean of the random samples
  for i in range(0,number_iter):
    np.random.seed(i)
    sample=np.random.choice(data,n)
    samples.append(sample)
    sample_means.append(np.mean(samples))
  pred_mean=np.mean(sample_means)
  return(sample_means,pred_mean)
```

```python
# Experiments for getting a suitable  sample size n for "Man group"
temp = [1,2,4,8,10,20,30,40,50]

for x in temp:
  plt.rcParams["figure.figsize"] = [12,8]
  sample_mean,means=bootstrap(int(len(df_man["Salary"])/x),1000,df_man['Salary']) 
  sample_mean=pd.DataFrame(data=sample_mean,columns=['Sample_Means'])
  
  # Plot histograms
  sample_mean.hist(bins=50,density=True)
  plt.title('n='+ str(1/x))
```

```python
# Experiments for getting a suitable  sample size n for "Woman group"
temp = [1,2,4,8,10,20,30,40,50]

for x in temp:
  plt.rcParams["figure.figsize"] = [12,8]
  sample_mean,means=bootstrap(int(len(df_woman["Salary"])/x),1000,df_woman['Salary']) 
  sample_mean=pd.DataFrame(data=sample_mean,columns=['Sample_Means'])
  
  # Plot histograms
  sample_mean.hist(bins=50,density=True)
  plt.title('n='+ str(1/x))
```

From the above graphs, I am taking sample size as 1/2 of the sample size of  both men and women group which produces a close enough normal distribution and has enough samples to consider.

Now I plot the bootstrapped data for men and women group together to compare.

```python
#bootstrapping data for man group with 1/2 sample size
samplesize_man= int(len(df_man)/2) #taking integer value 
mean_list_man=[]
#1000 repetitions
for i in range(1000):
  random.seed(i) #to ensure consistent results in every run
  sample = random.sample(df_man["Salary"].values.tolist(), samplesize_man)
  mean_list_man.append(statistics.mean(sample))

#bootstrapping data for woman group with 1/2 sample size
samplesize_woman= int(len(df_woman)/2) #taking integer value 
mean_list_woman=[]
#1000 repetitions
for i in range(1000):
  random.seed(i) #to ensure consistent results in every run
  sample_2 = random.sample(df_woman["Salary"].values.tolist(), samplesize_woman)
  mean_list_woman.append(statistics.mean(sample_2))

```

d. [0.5pts] If suitable, perform a two-sample t-test with a 0.05 threshold on the bootstrapped data. Explain your rationale.

Assumptions for a two-sample t test are:

1. Data should be continuous
2. The sample data should  have been randomly sampled from a population.
3. There is homogeneity of variance (i.e., the variability of the data in each group is similar).
4. The distribution is approximately normal.



From the graphs above we see that the data is continuous through out and we did sample from the entire population using random function from NumPy. Hence first two assumption are satisifed. We further saw visually from graphs as well as using shapiro test that both the data from men and women group are normally distributed. We also see the distribution of the difference in their means is also normal. Hence the fourth assumption also satisfies. Below we check for the assumption of homogenity of variance.

```python
#Checking homogenity if variance using Levene Test
stat,pval = stats.levene(mean_list_man,mean_list_woman,center= 'mean')
if pval>0.05:
    print('Homogeneity of Variance is not rejected')
else:
    print('Homogeneity of Variance is rejected')
```

Since there is no homogenity of variance, ideally normal t-test can't be performed. However, we can proceed with Welch's t-test that is designed for unequal population variances but the assumption of normality is maintained.

```python
# Welch's t test
t_stat,t_pval = stats.ttest_ind(mean_list_man, mean_list_woman,equal_var=False)
print ("t-test: t = %f  p = %f" % (t_stat, t_pval))
if t_pval >0.05:
    print('There is no statistically significant difference in the mean salary between men and women')
else:
    print('There is a statistically significant difference in the mean salary between men and women')
```
